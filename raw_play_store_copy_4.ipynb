### What did you know about your dataset?

In the 'Size' column, the values are represented as strings denoting size in various units such as 'M' for megabytes, 'k' for kilobytes, and 'Varies with devices.'

## **data cleaning and framing**


````python
# convert reviews type as integer type
play_store_df = play_store_df.astype({"Reviews": int})
````

````python
#drop $ sign from price column
play_store_df['Price'] = play_store_df['Price'].str.replace('$','')
````

````python
#convert price column type as float
play_store_df = play_store_df.astype({"Price": float})
````

````python
#in installs column we replace some sign and convert it into integer data type
play_store_df['Installs'] = play_store_df['Installs'].str.replace(',','')
play_store_df['Installs'] = play_store_df['Installs'].str.replace('+','')
play_store_df = play_store_df.astype({"Installs": int})
````

````python
#In size column apps size in mb and kb
play_store_df["Size"].unique()
````

Because the sizes of the applications in the dataset are given in both megabytes (MB) and kilobytes (KB), we converted the entire size column to megabytes to make it easier to work with the data.

````python
#After converted
play_store_df["Size"].unique()
````

````python
#data info after cleaning
play_store_df.info()
````

### Variables Description

**Dataset-1**

This set of data has information about different apps:

App: The name of the app and sometimes a short description.

Category: The type of app, with 33 different categories.

Size: How much space the app takes up when you install it.

Rating: The average score given by users, ranging from 1 to 5.

Reviews: The number of user reviews for the app.

Installs: An estimate of how many times the app has been installed.

Type: Whether the app is free or paid.

Price: The cost to install the app; it's zero for free apps.

Content rating: The age group for which the app is suitable.

Genres: The genre or genres to which the app belongs.

Last updated: The date when the latest update for the app was released.

Current Ver: The current version number of the app.

Android Ver: The version of the Android operating system the app is compatible with.

**Dataset-2**

This set of data includes:

App: The name of the app and sometimes a short description.

Translated_Review: The review given by a user, translated into English.

Sentiment: The emotional tone of the review, categorized as 'Positive,' 'Negative,' or 'Neutral.'

Sentiment_Polarity: A numerical value indicating whether the review is positive or negative.

Sentiment_Subjectivity: A value representing how closely the reviewer's opinion aligns with the general public's opinion, with higher values indicating closer alignment and lower values suggesting more factual information.
